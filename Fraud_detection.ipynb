```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
fraud_data = pd.read_csv("fraud_detection_bank_dataset.csv")
fraud_data.head()
```

```python
fraud_data.describe()
```

Checking if any of the columns contain any missing values

```python
fraud_data.isnull().sum()
```

There are no missing values. So we need not handle them. We now proceed to check if there are any duplicates.

```python
fraud_data.duplicated().sum()
```

There are no duplicates as well so we can now proceed any remove any unwanted columns.
The first column is Index value and it is not of any use for our analysis and hence it can be removed.

```python
X = fraud_data.drop(["Unnamed: 0"], axis=1)
```

We now check if any of the columns are correlated and delete columns that are highly correlated. 

```python
corr_df = X.corr()
for i in range(len(corr_df.columns)):
    correlated = [corr_df.iloc[:,i]>=0.9]
    if len(np.nonzero(correlated)[1])>1:
        print(np.nonzero(correlated)[1])
   
```

```python
X = X.drop(["col_7",'col_22',"col_40","col_47","col_49","col_50","col_55","col_57","col_63","col_64","col_65","col_66","col_56","col_54","col_60","col_64","col_62","col_50"], axis=1)
```

After removing the unnecessary columns, we now proceed to check if the dataset is balanced or imbalanced.

```python
X['targets'].value_counts()
```

As expected, the dataset is imbalanced and hence we balance the dataset by removing target-0 records.

```python
X_0 =X[X['targets']==0][:8500]
X_1 =X[X['targets']==1]
df = pd.concat([X_0,X_1],axis=0)
X = df.drop('targets',axis=1)
```

```python
target = df['targets']
target.value_counts()
```

Next we split the dataset into training and test set.
We further normalize the dataset.

```python
X_train, X_test, y_train, y_test = train_test_split(X, target, random_state = 0, test_size = 0.3)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

We first try the logistic regression model

```python
model = LogisticRegression()
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```

We then check the decision tree performance with our data

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
dt.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

Finally, we check with random forests

```python
rf = RandomForestClassifier(max_features=10, min_samples_split=5, min_samples_leaf=5,n_estimators= 100, criterion='gini')
#rf = RandomForestClassifier(max_features=5, min_samples_split=5, min_samples_leaf=5, oob_score=True,n_estimators= 300, criterion='entropy')
```

```python
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(rf.score(X_test, y_test))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

The random forest classifier worked best among all these models.
